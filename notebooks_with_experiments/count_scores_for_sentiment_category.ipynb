# Count scores for precomputed experiments results by GPT models

```python
import pandas as pd
from sklearn.metrics import balanced_accuracy_score
```

```python
#function to read data
def read_files_to_compare(path1, path2, column_name, column_number):
  df1=pd.read_csv(path1, sep=';')
  df2=pd.read_csv(path2, sep=';')
  df1_column=df1.iloc[column_number]
  df2_column=df2[column_name]
  return df1_column, df2_column

#function to count score
def count_score_on_columns(df1_column, df2_column, score_type='balanced_accuracy'):
  if score_type=='balanced_accuracy':
    score=balanced_accuracy_score(df1_column, df2_column)
  else:
    score=0
  return score
```

```python
def return_mean_score(name, path_to_file, col_name='Category'):
    score_sum=0
    for i in range(3):
        df1_column, df2_column = read_files_to_compare(path_to_file, path_gt, col_name, i)
        score_sum+=count_score_on_columns(df1_column, df2_column)
    print(name, score_sum/3)
    return name, score_sum/3
```

# Count confidence scores

For each file count average confidence for model and for human, when the prediction match and when they do not match.

```python
i=0
path_1='11 sierpnia first results of experiments/not_processed_1_model_3_5/category_notpreprocessed.csv'
path_1_conf='11 sierpnia first results of experiments/not_processed_1_model_3_5/category_notpreprocessed_confidence.csv'
col_name='Category'
df1_pred, df_gt_pred = read_files_to_compare(path_1, path_gt, col_name, i)
df1_conf, df_gt_conf = read_files_to_compare(path_1_conf, path_gt, 'Certainity_category', i)
```

```python
df1_pred, df_gt_pred
```

```python
conf_list_of_names_and_scores=[('Mean model same', 'Mean human same', 'Mean model not same', 'Mean human not same')]

def count_mean_confidence(name, path_1, path_1_conf, task_type='Category', certainity_type='Certainity_category'):
    count_same=0
    count_not_same=0
    sum_same_model=0
    sum_same_human=0
    sum_not_same_model=0
    sum_not_same_human=0
    for i in range(3):
        df1_pred, df_gt_pred = read_files_to_compare(path_1, path_gt, task_type, i)
        df1_conf, df_gt_conf = read_files_to_compare(path_1_conf, path_gt, certainity_type, i)
        for j in range(len(df1_pred)):
            if df1_pred[j]==df_gt_pred[j]:
                sum_same_model+=int(df1_conf[j])
                sum_same_human+=int(df_gt_conf[j])
                count_same+=1
            else:
                sum_not_same_model+=int(df1_conf[j])
                sum_not_same_human+=int(df_gt_conf[j])
                count_not_same+=1
    print(f"Mean confidence of model for same predictions: {sum_same_model/count_same}.")
    print(f"Mean confidence of human for same predictions: {sum_same_human/count_same}.")
    print(f"Mean confidence of model for not same predictions: {sum_not_same_model/count_not_same}.")
    print(f"Mean confidence of human for not same predictions: {sum_not_same_human/count_not_same}.")
    return name, sum_same_model/count_same, sum_same_human/ count_same, sum_not_same_model/count_not_same, sum_not_same_human/count_not_same
```
